```python
import requests
import pandas as pd
import polars as pl
# URL de la API

url =  "https://www.inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml/INDICATOR/6206824187,6207077561/es/0700/true/BISE/2.0/1a708388-cf7d-4d97-84f0-cb7a947e73f0?type=json"
rows = []

response = requests.get(url)

if response.status_code == 200:
    data = response.json()["Series"]

    for serie in data:
        for obs in serie["OBSERVATIONS"]:
            row = {
                "INDICADOR": serie["INDICADOR"],
                "FREQ": serie["FREQ"],
                "TOPIC": serie["TOPIC"],
                "UNIT": serie["UNIT"],
                "NOTE": serie.get("NOTE", ""),
                "SOURCE": serie.get("SOURCE", ""),
                "LASTUPDATE": serie["LASTUPDATE"],
                "TIME_PERIOD": obs["TIME_PERIOD"],
                "OBS_VALUE": obs["OBS_VALUE"],
                "OBS_STATUS": obs["OBS_STATUS"],
                "COBER_GEO": obs["COBER_GEO"]
            }
            rows.append(row)


    df = pl.DataFrame(rows)

    
    


else:
    print(f"Error en la solicitud: {response.status_code}")

df
```
